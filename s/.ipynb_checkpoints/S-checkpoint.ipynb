We can see from the above that the columns, “gender”, “class”, “test preparation course” are categorical variables.

Since marks are always numerical, therefore the columns,"science score" “math score”, “read score”, “write score” contain numerical values and hence are numerical variables.

Adding columns “total” and “average” to the dataset.

```python
df['total'] = df['science score'] + df['math score'] + df['reading score'] + df['writing score']
df['average'] = df['total'] / 4
```

```python
count_test = df['test'].value_counts()
labels = df['test'].value_counts().index
plt.figure(figsize= (6,6))
plt.pie(count_test,labels=labels,autopct='%1.1f%%')
plt.legend(labels)
plt.show()
```

```python
gender = {
    'male':1,
    'female':0
}
```

```python
df['gender']=df['gender'].map(gender)
df['Class']=df['Class'].map(Class)
df['test']=df['test'].map(test)
```

```python
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
#X = df.drop(columns=['average'])
X = df.drop(columns=['average'],axis=1)
#df.drop(['C', 'D'], axis = 1)
Y = df['average']
x_train, x_test, y_train, y_test = train_test_split(X, Y)
#test_size=0.30
```
